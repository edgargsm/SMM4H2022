### Imports

```python
import pandas as pd
import csv
from collections import Counter
import re
```

### Merging training data


```python
class_file = "/content/drive/MyDrive/Dissertacao/Subtask_1c/training_data/train_spans_norm.tsv"
tweet_file = "/content/drive/MyDrive/Dissertacao/Subtask_1c/training_data/train_tweets.tsv"

lowercase = False

# Read files
class_df = pd.read_csv(class_file, sep='\t', header=None)
tweet_df = pd.read_csv(tweet_file, sep='\t', quoting=csv.QUOTE_NONE, header=None)


tweet_dict = {}

for i in range(0, len(tweet_df)):
  text = tweet_df[1][i]
  if "&amp;" in text:
    text = text.replace("&amp;", "&")
  if lowercase:
    text = text.lower()
  tweet_dict[tweet_df[0][i]] = text

```

```python
with open("/content/drive/MyDrive/Dissertacao/Subtask_1c/training_data/merged_training_dataset.tsv", 'wt') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    tsv_writer.writerow(['text', 'ADE_label', 'start', 'end', 'span', 'label'])
    for i in range(len(class_df)):
      if "-" not in class_df[5][i]:
        tsv_writer.writerow([tweet_dict[class_df[0][i]], class_df[1][i], class_df[2][i], class_df[3][i], class_df[4][i], class_df[5][i]])
```

```python
# Unique Meddra UIDS / Terms
len(Counter(class_df[5]))
```

```python
train_id_set = set(Counter(class_df[5]).keys())
```

### Merging validation data


```python
class_file = "/content/drive/MyDrive/Dissertacao/Subtask_1c/validation_data/spans_norm.tsv"
tweet_file = "/content/drive/MyDrive/Dissertacao/Subtask_1c/validation_data/tweets.tsv"

lowercase = False

# Read files
class_df = pd.read_csv(class_file, sep='\t', header=None)
tweet_df = pd.read_csv(tweet_file, sep='\t', quoting=csv.QUOTE_NONE, header=None)


tweet_dict = {}

for i in range(0, len(tweet_df)):
  text = tweet_df[1][i]
  if "&amp;" in text:
    text = text.replace("&amp;", "&")
  if lowercase:
    text = text.lower()
  tweet_dict[tweet_df[0][i]] = text

```

```python
with open("/content/drive/MyDrive/Dissertacao/Subtask_1c/validation_data/merged_validation_dataset.tsv", 'wt') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    tsv_writer.writerow(['tweet_id', 'text', 'ADE_label', 'start', 'end', 'span', 'label'])
    for i in range(len(class_df)):
      tsv_writer.writerow([class_df[0][i], tweet_dict[class_df[0][i]], class_df[1][i], class_df[2][i], class_df[3][i], class_df[4][i], class_df[5][i]])
```

```python
# Unique Meddra UIDS / Terms
len(Counter(class_df[5]))
```

```python
val_id_set = set(Counter(class_df[5]).keys())
```

```python
intersection = 0
for med_id in list(val_id_set):
  if str(med_id) in train_id_set:
    intersection = intersection + 1

print("There are", intersection, "meddra terms in common between train and val sets.")
```
